```python
import requests
import re
from bs4 import BeautifulSoup
from bs4.element import Comment
from urllib.parse import urljoin
from IPython.display import display, HTML
```

```python
MAX_SENTENCE_LENGTH = 205 #sentence to fit one line

def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    #if isinstance(element, NavigableString):
    #    return False
    if element.name in ['nav', 'header', 'footer']:  # Ignore navbar, header, footer
        return False
    #if 'class' in element.attrs and any("navigation" in c or "button" in c or "nav" in c for c in element.attrs['class']):  # Ignore elements with specified classes
    #    return False
    #    classes_to_ignore = ['navigation', 'button', 'nav']
    #if 'class' in element.attrs and any(cls in element.get('class', []) for cls in classes_to_ignore):
    #    return False
    return True

def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)
    visible_text = u" ".join(t.strip() for t in visible_texts)
    
    # Split text into sentences using regular expressions
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', visible_text)
    
        # Split long sentences at the next non-alphabetical character
    for i, sentence in enumerate(sentences):
        if len(sentence) > MAX_SENTENCE_LENGTH:
            # Find the index of the next non-alphanumeric character after 213 characters
            split_index = MAX_SENTENCE_LENGTH
            while split_index < len(sentence) and sentence[split_index].isalnum():
                split_index += 1
            
            # Split the sentence
            if split_index < len(sentence):
                sentences[i] = sentence[:split_index+1] + '\n' + sentence[split_index+1:]
                
    return '\n'.join(sentences)

#def extract_text(url):
#Fetch the webpage content
#    response = requests.get(url)
#    body = response.text
    
    # Extract visible text
#    text = text_from_html(body)
    
#    return text

def extract_text(url, visited_urls=set()):
    if url in visited_urls:
        return ""
    
    # Fetch the webpage content
    response = requests.get(url)
    body = response.text
    
    # Extract visible text
    text = text_from_html(body)
    
    # Mark the URL as visited
    visited_urls.add(url)
    
    # Find all links on the page
    soup = BeautifulSoup(body, 'html.parser')
    links = soup.find_all('a', href=True)
    
    # Extract text from linked pages recursively
    for link in links:
        subpage_url = urljoin(url, link['href'])
        if subpage_url.startswith(url):
            text += extract_text(subpage_url, visited_urls)
    
    return text

def display_text_in_new_tab(text):
    # Generate HTML for displaying the text in a new tab
    html_content = f"""
    <script type="text/javascript">
        var newWindow = window.open("", "_blank");
        newWindow.document.write(`<pre>{text}</pre>`);
        newWindow.document.close();
    </script>
    """
    display(HTML(html_content))
```

```python
url = "https://medium.com/@sierraelman/the-future-of-poetry-26dabfc2f50a"
extracted_text = extract_text(url)
display_text_in_new_tab(extracted_text)
```
